# For beginners 入门者教程

```python
import numpy as np
import matplotlib.pylab as plt
```

##  Choose 选择适合自己的图形类型

### line plot 线图

```python
X = np.linspace(0,4*np.pi,1000)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X,Y)
fig.show()
```

### scatter plot 散点图

```python
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)

fig, ax = plt.subplots()
ax.scatter(X,Y)
fig.show()
```

### bar plot 条形图

```python
X = np.arange(10)
Y = np.random.uniform(0,10,10)

fig, ax = plt.subplots()
ax.bar(X,Y)
fig.show()
```

### image plot 栅格图

```python
Z = np.random.uniform(0,1,(8,8))

fig, ax = plt.subplots()
ax.imshow(Z)
fig.show()
```

### contourf plot 等值线

```python
Z = np.random.uniform(0,1,(8,8))

fig, ax = plt.subplots()
ax.contourf(Z)
fig.show()
```

### pie plot 饼图

```python
Z = np.random.uniform(0,1,4)

fig, ax = plt.subplots()
ax.pie(Z)
fig.show()
```

### hist plot 直方图

```python
Z = np.random.normal(0,1,100)

fig, ax = plt.subplots()
ax.hist(Z)
fig.show()
```

## Tweak 调整图形细节

### color 颜色

```python
X = np.linspace(0,4*np.pi,100)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X,Y,color='black')
fig.show()
```

### linestyle 线型

```python
X = np.linspace(0,4*np.pi,100)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X,Y,linestyle='--')
fig.show()
```

### linewidth 线宽

```python
X = np.linspace(0,4*np.pi,100)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X,Y,linewidth=5)
fig.show()
```

### marker 点型

```python
X = np.linspace(0,4*np.pi,100)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X,Y,marker='o')
fig.show()
```

## Organize 组织内容和子图

Both several data on one figure, and several subplots.
既可以多个数据画在一个图里，也可以组合子图

### several data in one 多个数据一张图

```python
X = np.linspace(0,4*np.pi,100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, ax = plt.subplots()
ax.plot(X,Y1,color='C1')
ax.plot(X,Y2,color='C2')
fig.show()
```

### subplots 组织子图

```python
X = np.linspace(0,4*np.pi,100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, axs = plt.subplots(2,1)

# fig, axs = plt.subplots((2,1)) Official website version, but not suitable for my version.这是官方给出的事例。但是我的版本下不能执行，我的版本下是上面那么写。 

axs[0].plot(X,Y1,color='C1')
axs[1].plot(X,Y2,color='C2')
fig.show()
```

```python
X = np.linspace(0,4*np.pi,100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, axs = plt.subplots(1,2)

# fig, axs = plt.subplots((2,1)) Official website version, but not suitable for my version.这是官方给出的事例。但是我的版本下不能执行，我的版本下是上面那么写。 

axs[0].plot(X,Y1,color='C1')
axs[1].plot(X,Y2,color='C2')
fig.show()
```

## Label 标题

```python
X = np.linspace(0,4*np.pi,100)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X,Y)
ax.set_title('A Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
fig.show()
```

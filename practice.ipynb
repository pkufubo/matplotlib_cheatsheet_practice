# For beginners 入门者教程

```python
import numpy as np
import matplotlib.pylab as plt
```

## line plot 线图

```python
X = np.linspace(0,4*np.pi,1000)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X,Y)
fig.show()
```

## scatter plot 散点图

```python
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)

fig, ax = plt.subplots()
ax.scatter(X,Y)
fig.show()
```

## bar plot 条形图

```python
X = np.arange(10)
Y = np.random.uniform(0,10,10)

fig, ax = plt.subplots()
ax.bar(X,Y)
fig.show()
```

## image plot 栅格图

```python
Z = np.random.uniform(0,1,(8,8))

fig, ax = plt.subplots()
ax.imshow(Z)
fig.show()
```
